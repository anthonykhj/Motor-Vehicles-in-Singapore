# Importing all libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Monthly Motor Vehicle Population by Vehicle Type (1 bar chart)

```python
filename = "monthly-motor-vehicle-population-by-vehicle-type.csv"
mvpop = np.genfromtxt(filename, dtype=("datetime64[Y]","U100",int), delimiter=",", names=True)
mvpop.dtype
```

```python
vtypes = np.unique(mvpop['vehicle_type'])
vtypes
```

```python
# Find average number of vehicles per vehicle type
avgs = []
labels = []
for v in vtypes:
    labels.append(v)
    avt = np.mean(mvpop[mvpop["vehicle_type"]==v]["number"])
    avgs.append(avt)
```

```python
plt.figure(figsize=(13, 9))
bar = plt.bar(labels,avgs,color="blue")
bar[0].set_color('yellow')
bar[2].set_color('black')
bar[3].set_color('green')
bar[4].set_color('m')
bar[5].set_color('lightblue')
bar[6].set_color('red')
bar[7].set_color('grey')
plt.title("No. of Vehicle by Vehicle Types", fontweight="bold")
plt.xlabel("Vehicle Types")
plt.xticks(rotation='vertical')
plt.ylabel("Average number of vehicles")
plt.show()
```
